Q1. Create a vehicle class with an __inhit__ method having instance variables as name _of_vehicle , max_speed and average_of_vehicle.

```python
Sure! Here's an example of a Python class called "Vehicle" with an __init__ method that initializes instance variables for name_of_vehicle, max_speed, and average_of_vehicle:

python

class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
In this class, the __init__ method is used to initialize the instance variables name_of_vehicle, max_speed, and average_of_vehicle. The self parameter represents the instance of the class being created, and the other parameters are used to set the initial values of the instance variables.

You can create an instance of the Vehicle class and access its instance variables like this:

python

car = Vehicle("Car", 200, 50)
print(car.name_of_vehicle)  # Output: Car
print(car.max_speed)  # Output: 200
print(car.average_of_vehicle)  # Output: 50
In this example, we create an instance of the Vehicle class called car with the name "Car", a max speed of 200, and an average speed of 50. We then print the values of the instance variables using dot notation.
```

Q3. What is multiple inheritance?Write a python code to demonstrate multiple inhertance.

```python
Multiple inheritance is a feature in object-oriented programming languages where a class can inherit attributes and methods from multiple parent classes. This allows a derived class to inherit and combine the properties of multiple classes.

Here's an example of a Python code that demonstrates multiple inheritance:

python

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subclass must implement this method")


class Mammal(Animal):
    def speak(self):
        return "Mammal: Roar!"


class Dog(Mammal):
    def speak(self):
        return "Dog: Woof!"


class Cat(Mammal):
    def speak(self):
        return "Cat: Meow!"


class Bat(Mammal):
    def speak(self):
        return "Bat: Squeak!"


class Bird(Animal):
    def speak(self):
        return "Bird: Chirp!"


class Parrot(Bird):
    def speak(self):
        return "Parrot: Hello!"


class Chimera(Dog, Cat):
    pass


# Create instances and invoke methods
chimera = Chimera("Chimera")
print(chimera.speak())
In this example, we have several classes representing animals. The Animal class is the base class with a basic speak method that raises a NotImplementedError because it should be overridden by subclasses.

The Mammal, Bird, Dog, Cat, Bat, and Parrot classes are derived from the Animal class and each overrides the speak method to provide their own specific implementation.

The Chimera class is a derived class that demonstrates multiple inheritance. It inherits from both the Dog and Cat classes. Since it doesn't have its own speak method, it inherits the speak method from both parent classes.

Finally, we create an instance of Chimera called chimera and invoke its speak method. The output will be "Dog: Woof!" because the Chimera class inherits the speak method from the Dog class, which is the first parent class in the multiple inheritance hierarchy.




```

Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this class.

```python
In Python, getters and setters are methods used to access and modify the values of private instance variables of a class. They provide a controlled way of accessing and updating the values of class attributes.

Here's an example of a class with a getter and a setter method in Python:

python

class Person:
    def __init__(self, name):
        self._name = name  # private instance variable with a single underscore prefix

    # Getter method
    def get_name(self):
        return self._name

    # Setter method
    def set_name(self, name):
        self._name = name
In this example, the Person class has a private instance variable _name, indicated by a single underscore prefix. The getter method get_name allows external code to retrieve the value of _name, and the setter method set_name allows external code to update the value of _name.

Here's an example of how you can use the getter and setter methods:

python

person = Person("Alice")
print(person.get_name())  # Output: Alice

person.set_name("Bob")
print(person.get_name())  # Output: Bob
In this example, we create an instance of the Person class called person with the name "Alice". We then use the getter method get_name to retrieve the value of the private instance variable _name and print it. Afterward, we use the setter method set_name to update the value of _name to "Bob" and again use the getter method to retrieve and print the updated value.




```

Q5. What is method overriding in python?Write a python code to demonstrate method overriding.

```python
Method overriding is a feature in object-oriented programming where a subclass provides a different implementation for a method that is already defined in its superclass. It allows the subclass to change the behavior of the inherited method according to its specific requirements.

Here's an example of a Python code that demonstrates method overriding:

python

class Shape:
    def area(self):
        raise NotImplementedError("Subclass must implement this method")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


# Create instances and invoke methods
rectangle = Rectangle(4, 5)
print("Area of rectangle:", rectangle.area())  # Output: 20

circle = Circle(3)
print("Area of circle:", circle.area())  # Output: 28.26
In this example, we have a base class called Shape with an area method that raises a NotImplementedError because it should be overridden by subclasses.

The Rectangle class is derived from Shape and overrides the area method to provide its own implementation. It calculates the area by multiplying the length and width.

The Circle class is also derived from Shape and overrides the area method. It calculates the area by multiplying the square of the radius with the constant value of 3.14.

We create instances of Rectangle and Circle, and then invoke their area methods. The output will be the calculated areas specific to each shape, demonstrating method overriding.




```
